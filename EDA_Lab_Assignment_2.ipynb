### **5. Try replacing all the NaN's in the sf_permits data with the one that comes**

  > a. directly after it and then replacing any remaining NaN's with 0 (Backward Fill)
  
  > b. Also try with most recent preceding (previous) non-missing value in the same column. (Forward Fill)

```python
backward_fill = sf_permits.fillna(method="bfill",axis=0,limit=1).fillna(0)
backward_fill
```

```python
backward_fill.isnull().sum()
```

```python
sf_permits.head(20)
```
